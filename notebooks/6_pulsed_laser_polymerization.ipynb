# Pulsed-Laser Polymerization

## Theory

Pulsed-laser polymerization coupled with size-exclusion chromatography (PLP-SEC) is the recommended method for experimentally determining the propagation rate coefficients ($k_p$) in radical polymerization. This method, originally developed by Olaj and co-workers in the late 1980s, involves periodically irradiating a monomer solution containing a small amount of photoinitiator with laser pulses. When experimental conditions are optimized, the molar mass distribution of the resulting polymer exhibits a characteristic pattern from which the $k_p$ value can be inferred.

The governing equations for this system largely mirror those of a standard transient radical polymerization process (refer to [Notebook 5](5_radical_polymerization.ipynb)). The key difference lies in the initiation step. Following the works of [Hutchinson et al.](https://doi.org/10.1021/ma00094a016) and [Nikitin et al.](https://doi.org/10.1002/mats.200350005), the reaction scheme can be summarized as:

$$ I \xrightarrow{h\nu} 2 I^{\cdot} $$

$$ I^{\cdot} + M \xrightarrow{k_i} R_1^{\cdot} $$

$$ R_n^{\cdot} + M \xrightarrow{k_p} R_{n+1}^{\cdot} $$

$$ R_n^{\cdot} + M \xrightarrow{k_{tr,M}} P_{n} + R_1^{\cdot} $$

$$ R_n^{\cdot} + R_m^{\cdot} \xrightarrow{k_{tc}} P_{n+m} $$

$$ R_n^{\cdot} + R_m^{\cdot} \xrightarrow{k_{td}} P_n + P_m $$

$$ R_n^{\cdot} + I^{\cdot} \xrightarrow{k_{t}} P_n $$

$$ I^{\cdot} + I^{\cdot} \xrightarrow{k_{t}} dead $$

Assuming constant rate coefficients (i.e., no chain-length dependence), the transient species balances can be expressed as:

\begin{align*}
\frac{\textrm{d}[I^{\cdot}]}{\textrm{d}t}   & = [\Delta I^{\cdot}]\sum_{j} \delta(t-j t_p) - \left( k_i[M] + k_t [T^{\cdot}] \right)[I^{\cdot}] \\
\frac{\textrm{d}[M]}{\textrm{d}t}           & = - \left( k_i [I^{\cdot}] + (k_p + k_{tr,M})[R^{\cdot}] \right)[M] \\
\frac{\textrm{d}[R_1^{\cdot}]}{\textrm{d}t} & = k_i[M][I^{\cdot}] + k_{tr,M}[M][R^{\cdot}] - \left((k_p + k_{tr,M})[M] + k_t [T^{\cdot}]\right)[R_1^{\cdot}]  \\
\frac{\textrm{d}[R_n^{\cdot}]}{\textrm{d}t} & = k_p[M][R_{n-1}^{\cdot}] - \left((k_p + k_{tr,M})[M] + k_t [T^{\cdot}]\right)[R_n^{\cdot}], \; n \ge 2 \\
\frac{\textrm{d}[P_n]}{\textrm{d}t} & = \left(  k_{tr,M} [M] + k_{td} [R^{\cdot}] + k_t[I^{\cdot}]\right)[R_n^{\cdot}]
+ \tfrac{1}{2} k_{tc} \sum_{j=1}^{n-1}[R_{n-j}^{\cdot}][R_j^{\cdot}] \\
\frac{\textrm{d}[R^{\cdot}]}{\textrm{d}t}   & =  k_i[M][I^{\cdot}]  - k_t [T^{\cdot}][R^{\cdot}] \\
\frac{\textrm{d}[P]}{\textrm{d}t}   & =  \left( k_{tr,M} [M] + k_{td} [R^{\cdot}] + k_t[I^{\cdot}] \right)[R^{\cdot}] + \tfrac{1}{2} k_{tc} [R^{\cdot}]^2
\end{align*}

Here, $[\Delta I^{\cdot}]$ represents the burst of primary radicals generated immediately following each laser pulse, and $t_p$ refers to the time interval between successive laser pulses. The last two equations describe the evolution of the zeroth moments of $R^{\cdot}$ (the total radical concentration) and $P$ (the total polymer concentration), where $[R^{\cdot}] = \sum_n [R_n^{\cdot}]$ and $[P] = \sum_n [P_n]$. Note: these equations do _not_ respect the $2k_t$ convention.

By defining an appropriate upper limit for the chain length, $N$, we obtain a closed system of $2N + 4$ ordinary differential equations (ODEs). While, as discussed in [Notebook 5](5_radical_polymerization.ipynb), directly integrating such an ODE system may not be the most computationally efficient approach, it is straightforward to implement.

## Numerical Solution

```python
import matplotlib.pyplot as plt
import numpy as np
from numba import njit
from scipy.integrate import solve_ivp
```

### Model Equations

First, we define a function to compute the time derivative of the ODE system, i.e., $\dot{y}(t, y)$. To enhance performance, this function is jitted using [Numba](https://numba.pydata.org/) â€” every bit of optimization helps when dealing with such complex systems.

```python
@njit(fastmath=True)
def model_ydot(t: float,
               y: np.ndarray,
               ki: float,
               kp: float,
               ktc: float,
               ktd: float,
               kfm: float,
               N: int
               ) -> np.ndarray:
    """Calculate derivative of the state vector, dy/dt.

    y = [I, M, r0, p0, R1..RN, P1..PN]

    r0 = sum(Rn)
    p0 = sum(Pn)
    
    Parameters
    ----------
    t : float
        Time.
    y : np.ndarray
        State vector.
    ki : float
        Initiation rate coeficient.
    kp : float
        Propagation rate coeficient.
    ktc : float
        Termination by combination rate coeficient.
    ktd : float
        Termination by disproportionation rate coeficient.
    kfm : float
        Transfer to monomer rate coeficient.
    N : int
        Maximum chain length.

    Returns
    -------
    np.ndarray
        Time derivative of the state vector.
    """
    # Unpack the state vector
    I = y[0]
    M = y[1]
    r0 = y[2]
    # p0 = y[3]
    R = y[4:N+4]
    # P = y[N+4:]

    # Total radical concentration
    T = I + r0
    kt = ktc + ktd

    ydot = np.empty_like(y)
    Idot = ydot[0:1]
    Mdot = ydot[1:2]
    r0dot = ydot[2:3]
    p0dot = ydot[3:4]
    Rdot = ydot[4:N+4]
    Pdot = ydot[N+4:]

    Idot[:] = -(ki*M + kt*T)*I
    Mdot[:] = -(ki*I + (kp + kfm)*r0)*M
    r0dot[:] = ki*I*M - kt*T*r0
    p0dot[:] = (kfm*M + ktd*r0 + kt*I)*r0 + 0.5*r0**2
    Rdot[:] = -((kp + kfm)*M + kt*T)*R
    Rdot[0] += (ki*I + kfm*r0)*M
    Rdot[1:] += kp*M*R[:-1]
    if ktc > 1e0:
        Pdot[:] = 0.
        # Pdot[:] = 0.5*ktc*np.convolve(R, R, mode='full')[:R.size]
    else:
        Pdot[:] = 0.
    Pdot[:] += (kfm*M + ktd*r0 + kt*I)*R

    return ydot
```

Then, we perform the numerical integration using a suitable ODE solver. This system is _very_ [stiff](https://en.wikipedia.org/wiki/Stiff_equation), therefore we need to chose an implicit solver like `LSODA`.

```python
def solve_model(M0: float,
                dI: float,
                ki: float,
                kp: float,
                kt: float,
                alpha: float,
                CtrM: float,
                N: int,
                prr: float,
                npulses: int,
                ) -> tuple[np.ndarray, ...]:
    """Solve the dynamic model.

    Parameters
    ----------
    M0 : float
        Initial monomer concentration.
    dI : float
        Primary radical burst.
    ki : float
        Initiation rate coeficient.
    kp : float
        Propagation rate coefficient.
    kt : float
        Termination rate coefficient.
    alpha : float
        Relative contribution of disportionation, ktd/kt.
    CtrM : float
        Ratio of transfer to monomer, ktrM/kp.
    N : int
        Maximum chain length.
    prr : float
        Pulse repetition rate.
    npulses : int
        Number of pulses.

    Returns
    -------
    tuple[np.ndarray, ...]
        idx_pulses, t, s, I, M, r0, p0, R, P
    """

    # Chain lengths
    s = np.arange(1, N+1)

    # Initial conditions
    R0 = np.zeros_like(s)
    P0 = np.zeros_like(s)
    y0 = np.concatenate(([0.+dI, M0, R0.sum(), P0.sum()], R0, P0))

    # Rate coefficients
    kfm = kp*CtrM
    ktd = kt*alpha
    ktc = kt*(1 - alpha)

    # PLP period
    tp = 1/prr

    # Absolute tolerances (critically important!)
    atol = np.concatenate(
        ([1e-13, 1e0, 1e-9, 1e-6], 1e-13*np.ones(N), 1e-10*np.ones(N)))

    # Solve ODE set for each pulse
    t0 = 0.
    solutions = []
    idx_pulses = []
    for _ in range(npulses):
        solution = solve_ivp(model_ydot,
                             t_span=(t0, t0+tp),
                             y0=y0,
                             args=(ki, kp, ktc, ktd, kfm, N),
                             method='LSODA',  # implicit solver is a MUST
                             rtol=1e-4,
                             atol=atol)

        solutions.append(solution)
        idx_pulses.append(solution.t.size)

        # Update initial conditions for the next pulse
        t0 = solution.t[-1]
        y0 = solution.y[:, -1].copy()
        y0[y0 < 1e-15] = 0.  # discard numerical noise
        y0[0] += dI         # add primary radical burst

    # Unpack the solution
    idx_pulses = np.cumsum(idx_pulses) - 1
    t = np.concatenate([solution.t for solution in solutions])
    y = np.hstack([solution.y for solution in solutions])
    I = y[0, :]
    M = y[1, :]
    r0 = y[2, :]
    p0 = y[3, :]
    R = y[4:N+4, :]
    P = y[N+4:, :]

    return idx_pulses, t, s, I, M, r0, p0, R, P
```

### Input Parameters

We will use the parameters for bulk methyl methacrylate (MMA) at 40Â°C, as reported by [Hutchinson et al.](https://doi.org/10.1021/ma00094a016).

```python
# Monomer concentration (bulk)
MM = 100.121 # g/mol
rho = 940.   # g/L
M0 = rho/MM  # mol/L

# Rate coefficients
kp = 480    # L/(mol.s)
ki = 2*kp
kt = 5e7    # L/(mol.s)
alpha = 1 
CtrM = 8e-6

# Laser and photoinitator parameters
prr = 10     # Hz
dI = 0.5e-6  # mol/L

# Numerical settins
npulses = 5
N = 4000
```

### ODE Solution

Finally, we perform the integration. The duration of this operation depends on the size of $N$ and the time scale of the fastest reaction, which is typically the propagation step.

```python
idx_pulses, t, s, I, M, r0, p0, R, P = solve_model(M0, dI, ki, kp, kt, alpha, CtrM, N, prr, npulses)
```

### Plots

#### Reactant and Polymer (0-th moment) Concentrations

```python
fig1, ax = plt.subplots(4, 1, sharex=True, figsize=(8, 8))

fig1.suptitle("Evolution of species concentrations".title())

# Add trace for [I*]
ax[0].plot(t, I)
ax[0].set_ylabel(r"$[I^{\cdot}]$" + " (mol/L)")
ax[0].grid(True)

# Add trace for [M]
ax[1].plot(t, M)
ax[1].set_ylabel(r"$[M]$" + " (mol/L)")
ax[1].grid(True)

# Add trace for [R*]
ax[2].plot(t, r0, label='moment')
ax[2].plot(t, R.sum(axis=0), linestyle='--', label='sum(Rn)')
ax[2].set_ylabel(r"$[R^{\cdot}]$" + " (mol/L)")
ax[2].grid(True)
ax[2].legend()

# Add trace for [P]
ax[3].plot(t, p0, label='moment')
ax[3].plot(t, P.sum(axis=0), linestyle='--', label='sum(Pn)')
ax[3].set_ylabel(r"$[P]$" + " (mol/L)")
ax[3].grid(True)
ax[3].legend(loc='lower right')

ax[-1].set_xlabel("Time (s)")
```

#### Chain Length Distribution

```python
fig2, ax = plt.subplots(3, 1, sharex=True, figsize=(8, 8))
fig2.suptitle("Distributions At End of Each Period")
fig2.tight_layout()

# Number distribution
CLDn = P/(p0[np.newaxis, :] + 1e-50)

# GPC distribution normalized by peak
CLDgpc = (s[:, np.newaxis]**2)*CLDn
CLDgpc /= (CLDgpc.max(axis=0)[np.newaxis, :] + 1e-50)

for pulse, idx in enumerate(idx_pulses):
    label = f"{pulse+1}"
    ax[0].plot(s*MM, R[:, idx], label=label)
    ax[1].plot(s*MM, P[:, idx])
    ax[2].plot(s*MM, CLDgpc[:, idx])

ax[0].legend()
ax[0].grid(True)
ax[1].grid(True)
ax[2].grid(True)
ax[0].set_ylabel(r"$[R_n^{\cdot}]$" + " (mol/L)")
ax[1].set_ylabel(r"$[P_n]$" + " (mol/L)")
ax[2].set_ylabel(r"$w(\log(M))$")
ax[-1].set_xscale('log')
ax[-1].set_xlabel("Molar mass [g/mol]")
```
